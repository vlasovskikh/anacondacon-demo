
## Query stocks data from an SQL data source

Let's explore the stocks table here:


```python


import pandas as pd
import sqlite3

with sqlite3.connect('data_source.db') as conn:
    stocks = pd.read_sql_query("""
        SELECT * FROM stocks
        ORDER BY price
        """, conn)

stocks
```


## Visualize a 2D normal distribution


```python

import holoviews as hv
import numpy as np
from holoviews import opts

hv.extension('bokeh')
hv.renderer('bokeh').theme = 'dark_minimal'
hv.output(fig='png')

np.random.seed(44)
data = np.random.randn(10000, 2)
```

```python

data
```

```python

hex_tiles = hv.HexTiles(data)
hex_tiles.opts(opts.HexTiles(cmap='plasma',
                             width=500, height=400,
                             colorbar=True))

```
